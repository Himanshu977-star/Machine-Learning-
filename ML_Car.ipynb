```python
data=data[~(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)]
```

```python
data.shape
```

```python
sorted(data['Selling_Price'],reverse=True)
```

```python
data.head()
```

```python
data['Fuel_Type'].unique()
```

```python
data['Fuel_Type']=data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
data.head()
```

```python
data['Transmission'].unique()
```

```python
data['Transmission']=data['Transmission'].map({'Manual':0, 'Automatic':1})
```

```python
data.head()
```

```python
data['Seller_Type'].unique()
```

```python
data['Seller_Type']=data['Seller_Type'].map({'Dealer':0, 'Individual':1})
```

```python
data.head()
```

```python
x=data.drop(['Car_Name','Selling_Price'],axis=1)
y=data['Selling_Price']
```

```python
data.head()
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)

rf=RandomForestRegressor()
rf.fit(X_train,y_train)

xgb=GradientBoostingRegressor()
xgb.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lr,X_test,y_test)
```

```python
cross_val_score(rf,X_test,y_test)
```

```python
cross_val_score(xgb,X_test,y_test)
```

```python
lr_avg=cross_val_score(lr,X_test,y_test).mean()
rf_avg=cross_val_score(rf,X_test,y_test).mean()
xgb_avg=cross_val_score(xgb,X_test,y_test).mean()
```

```python
lr_avg
```

```python
rf_avg
```

```python
xgb_avg
```

```python
x_pred1=lr.predict(X_test)
x_pred2=rf.predict(X_test)
x_pred3=xgb.predict(X_test)
```

```python
from sklearn import metrics
```

```python
score1=metrics.r2_score(y_test,x_pred1)
score2=metrics.r2_score(y_test,x_pred2)
score3=metrics.r2_score(y_test,x_pred3)
```
